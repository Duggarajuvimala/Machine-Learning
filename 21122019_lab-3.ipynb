### 8. Apply MinMaxScaler() and StandardScaler() to the following features: calories, protien, fat, sodium, fiber, carbo, sugars.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(data[['calories','protein','fat','sodium','fiber','carbo','sugars']])
print(scaled)
```

```python
from pandas import DataFrame
from matplotlib import pyplot
dataset = DataFrame(scaled)
# summarize
print(dataset.describe())
# histograms of the variables
dataset.hist()
pyplot.show()
```

```python
fig = plt.figure(figsize=(15,7))
plt.hist(scaled, bins = 15)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled1 = scaler.fit_transform(data[['calories','protein','fat','sodium','fiber','carbo','sugars']])
print(scaled1)
```

```python
from pandas import DataFrame
from matplotlib import pyplot
dataset = DataFrame(scaled1)
# summarize
print(dataset.describe())
# histograms of the variables
dataset.hist()
pyplot.show()
```

- **MinMaxScaler()**
 - which is used for Data Normalization that is 'Your data doesnâ€™t follow Normal/ Gaussian distribution'
   - Data normalization, in this case, is the process of rescaling one or more attributes to the range of 0 to 1. This means that the largest value for each attribute is 1 and the smallest value is 0.
   - that we can see in the first plot where x-axes ranges from (0.0,0.1)
- **StandardScaler()**
 - Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.
  - where we can see that the ranges from (-4,4)
